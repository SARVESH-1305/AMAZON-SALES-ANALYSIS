# Amazon Sales Data Analysis
This Jupyter Notebook performs data analysis on Amazon sales data, including:
- ETL Process (Extract, Transform, Load)
- Sales trends analysis (Year-wise, Month-wise, Year-Month-wise)
- Key insights: Top products, Regional sales distribution
- Visualizations using Matplotlib & Seaborn

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_excel('Amazon Sales data.xlsx')
df

# Convert 'Order Date' to datetime format
df["Order Date"] = pd.to_datetime(df["Order Date"])

# Extract year, month, and year-month for trend analysis
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month
df["Year_Month"] = df["Order Date"].dt.to_period("M")

df.head()
```

```python
# Yearly Sales Trend
yearly_sales = df.groupby("Year")["Total Revenue"].sum()

plt.figure(figsize=(10, 5))
sns.lineplot(x=yearly_sales.index, y=yearly_sales.values, marker="o", color="b")
plt.title("Yearly Sales Trend (Total Revenue)")
plt.xlabel("Year")
plt.ylabel("Total Revenue")
plt.grid(True)
plt.show()
```

```python
# Month-wise Sales Trend
monthly_sales = df.groupby("Month")["Total Revenue"].sum()

plt.figure(figsize=(10, 5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker="o", color="g")
plt.title("Month-wise Sales Trend (Total Revenue)")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

```python
# Year-Month-wise Sales Trend
year_month_sales = df.groupby("Year_Month")["Total Revenue"].sum()

plt.figure(figsize=(12, 5))
sns.lineplot(x=year_month_sales.index.astype(str), y=year_month_sales.values, marker="o", color="r")
plt.title("Year-Month-wise Sales Trend (Total Revenue)")
plt.xlabel("Year-Month")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Top 10 Selling Products
top_products = df.groupby("Item Type")["Total Revenue"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=top_products.values, y=top_products.index, palette="viridis")
plt.title("Top 10 Best-Selling Products by Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Product Type")
plt.grid(axis="x")
plt.show()
```

```python
# Sales by Region
region_sales = df.groupby("Region")["Total Revenue"].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 5))
sns.barplot(x=region_sales.values, y=region_sales.index, palette="coolwarm")
plt.title("Sales Distribution by Region")
plt.xlabel("Total Revenue")
plt.ylabel("Region")
plt.grid(axis="x")
plt.show()
```

 # KEY METRICS CALCULATION 

```python
# Total Sales
total_sales = df['Total Revenue'].sum()
print("Total Sales: $",total_sales)

# Average Order Value
avg_order_value = df['Total Revenue'].sum() / df['Order ID'].nunique()
print("Average Order Value: $",avg_order_value)

# Most Profitable Products
top_products = df.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False)
print(top_products.head())
```

# Correlation Analysis  

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df[['Units Sold', 'Unit Price', 'Total Revenue', 'Total Profit']].corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
